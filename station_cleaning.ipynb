```python
import pandas as pd
import numpy as np
```

```python
bike_df = pd.read_csv("202210-citibike-tripdata.csv")
bike_df
```

```python
station_start_df = bike_df[["start_station_name", "start_lat", "start_lng"]]
station_start_df.dropna()
```

```python
station_start_df = station_start_df.drop_duplicates(subset=["start_station_name"])
station_start_df
```

```python
station_start_df.to_csv("start-stations.csv", index=False)
```

```python
station_end_df = bike_df[["end_station_name", "end_lat", "end_lng"]]
station_end_df.dropna()
```

```python
station_end_df = station_end_df.drop_duplicates(subset=["end_station_name"])
station_end_df
```

```python
station_end_df.to_csv("end-stations.csv", index=False)
```
